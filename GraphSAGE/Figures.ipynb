```python
zero_auroc = [0.50,0.51,0.51,0.50,0.52]
zero_auprc = [0.04,0.04,0.044,0.04,0.07]
one_auroc = [0.55,0.57,0.53,0.55,0.58]
one_auprc = [0.32,0.34,0.32,0.31,0.33]
high_auroc = [0.72,0.73,0.71,0.73,0.72]
high_auprc = [0.69,0.70,0.68,0.70,0.69]
```

```python
plt.figure(dpi=300)
categories = ['','','']#['Low Degree', 'High Degree']
values1 = [np.mean(zero_auroc), np.mean(one_auroc),np.mean(high_auprc)]
values2 = [np.mean(zero_auprc), np.mean(one_auprc),np.mean(high_auprc)]

errors1 = [np.std(zero_auroc), np.std(one_auroc),np.std(high_auprc)]
errors2 = [np.std(zero_auprc), np.std(one_auprc),np.std(high_auprc)]

bar_width = 0.2

x = np.arange(len(categories))

plt.bar(x, values1, bar_width, yerr=errors1, label='AUROC',alpha=0.3)
plt.bar(x + bar_width, values2, bar_width, yerr=errors2, label='AUPRC', alpha=0.3)

# Adding labels, title, and legend
#plt.ylabel('Performance')
plt.xticks(x + bar_width/2, categories)
plt.legend(prop={'size': 10})

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```
