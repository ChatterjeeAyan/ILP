# Combined Plots


```python
df = pd.DataFrame()
df['attributes'] = ['Random'] * 3 + ['Shuffled'] * 3 + ['Pre-trained'] * 3
df['dataset'] = ['ogbl-ppa','ogbl-collab','ogbl-ddi'] * 3
df['perf'] = [0.49, 0.49, 0.49, 0.50, 0.64, 0.50, 0.75, 0.98, 0.51]
df['error'] = [0.03,0.02,0.02,0.04,0.02,0.02,0.02,0.03,0.02]

plt.figure()
ax = sns.barplot(x='attributes', y='perf', hue='dataset', data=df)
x_coords = [p.get_x() + 0.5 * p.get_width() for p in ax.patches]
y_coords = [p.get_height() for p in ax.patches]
ax.errorbar(x=x_coords, y=y_coords, yerr=df["error"], fmt="none", c="k")
plt.show()
         
#sns.barplot(x='attributes', y='perf', hue='dataset', data=df, yerr=df['error'])
```
